Import libraries and assets

```python
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline
```

```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)


URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), "opencv_bootcamp_assets_NB3.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)   
```

Let's first see the image we got:

```python
apollo = cv2.imread("Apollo_11_Launch.jpg", 1)
plt.imshow(apollo)
```

A line can be drawn by using `cv2.line(img, pt1, pt2, color, thickness, lineType)`

where:
- img is the image duhh
- pt1 is a 2-tuple of the coordinates (from mpl coords) of where the line starts
- pt2 is a 2-tuple of the coordinates (from mpl coords) of where the line ends
- color is the 3-tuple rgb code for what color you want
- thickness (optional) is an integer of how thick you want the line to be
- lineType is the type of line (type `cv2.line_` to see options)

```python
apollo_line = apollo.copy()

cv2.line(apollo_line, (200,100), (1200,700), (255, 0, 0), 5, cv2.LINE_AA)

plt.imshow(apollo_line)
```

A circle can be drawn using `cv2.circle(img, center, radius, color, thickness, lineType)`

where:
- img is the image
- center is the 2-tuple coordinates of where the center is
- radius is how many pixels from the center you want
- color is the rgb 3-tuple
- thickness is how thick you want the line to be (put -1 to fill circle)
- lineType is the type of line

```python
apollo_circle = apollo.copy()

cv2.circle(apollo_circle, (600,350), 300, (0, 255, 0), 3)
plt.imshow(apollo_circle)
```

A rectangle can be drawn using `cv2.rectangle(img, pt1, pt2, color, thickness, lineType)`

where:
- img is the image
- pt1 is a 2-tuple of the coordinates (from mpl coords) of the bottom left corner of the rectangle
- pt2 is a 2-tuple of the coordinates (from mpl coords) of the top right corner of the rectangle
- color is the rgb 3-tuple
- thickness is how thick you want the line to be (put -1 to fill rectangle)
- lineType is the type of line

```python
apollo_rect = apollo.copy()

cv2.rectangle(apollo_rect, (420,650), (750, 50), (0, 0, 255), 2)
plt.imshow(apollo_rect)
```

Text can be added through `cv2.putText(img, text, org, fontFace, fontScale, color, thickness, lineType)`

where:
- img is the image
- text is the message you want to put
- org is the coordinates of the bottom left corner of the text on where to put it
- fontFace is the type of font (type `cv2.font` to see options)
- fontScale is the size scaling based on the fontFace (put -1 to flip text)
- color is the rgb 3-tuple
- thickness is how thick you want the line to be
- lineType is the type of line

```python
apollo_text = apollo.copy()

cv2.putText(apollo_text, "The world before rapture", (250,700), cv2.FONT_HERSHEY_DUPLEX, 2, (255, 0, 0), 3)
plt.imshow(apollo_text)
```
